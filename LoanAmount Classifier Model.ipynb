```python
df["Gender"].fillna("Male",inplace=True)
```

```python
df["Married"].fillna("Yes",inplace=True)
```

```python
df["Dependents"].fillna(0,inplace=True)
```

```python
df["Education"].fillna("Graduate",inplace=True)
```

```python
df["Self_Employed"].fillna("No",inplace=True)
```

```python
df["LoanAmount"].fillna(df["LoanAmount"].mean(skipna=True),inplace=True)
```

```python
df["Loan_Amount_Term"].fillna(360,inplace=True)
df["Credit_History"].fillna(1,inplace=True)
```

```python
df.head(12)
```

```python
df.isna().sum().sum()
```

```python
df['ApplicantIncome']=df['ApplicantIncome'].astype(int)
```

```python
df.info()
```

```python
df=df.drop(['Loan_ID'],axis=1)
```

```python
df['Dependents']=df['Dependents'].replace("3+","3")
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
a=['Gender']
b=['Married']
d=['Education']
e=['Self_Employed']
f=['Property_Area']
encoder=OrdinalEncoder()
df[a]=encoder.fit_transform(df[a])
df[b]=encoder.fit_transform(df[b])
df[d]=encoder.fit_transform(df[d])
df[e]=encoder.fit_transform(df[e])
df[f]=encoder.fit_transform(df[f])
df.head(9)
```

```python
df=df.astype({"CoapplicantIncome":'int',"LoanAmount":'int',"Loan_Amount_Term":'int',"Credit_History":'int',"Gender":'int',"Married":'int',"Dependents":'int',"Education":'int',"Self_Employed":'int',"Property_Area":'int'})
display(df)
```

```python
df.info()
```

```python
corr=df.corr()
corr.style.background_gradient()
```

```python
x=df[["Gender","Married","Dependents","Education","Self_Employed","Property_Area","ApplicantIncome","CoapplicantIncome","LoanAmount","Loan_Amount_Term","Credit_History"]]
y=df["Loan_Status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=27)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```
