```python
dt['Dependents']=dt['Dependents'].replace("3+","3")
```

```python
dt["Gender"].fillna("Male",inplace=True)
dt["Dependents"].fillna(0,inplace=True)
dt["Self_Employed"].fillna("No",inplace=True)
dt["LoanAmount"].fillna(dt["LoanAmount"].mean(skipna=True),inplace=True)
dt["Loan_Amount_Term"].fillna(360,inplace=True)
dt["Credit_History"].fillna(1,inplace=True)
```

```python
dt=dt.drop(['Loan_ID'],axis=1)
```

```python
dt.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
a=['Gender']
b=['Married']
d=['Education']
e=['Self_Employed']
f=['Property_Area']
encode=OrdinalEncoder()
dt[a]=encode.fit_transform(dt[a])
dt[b]=encode.fit_transform(dt[b])
dt[d]=encode.fit_transform(dt[d])
dt[e]=encode.fit_transform(dt[e])
dt[f]=encode.fit_transform(dt[f])
```

```python
dt.head()
```

```python
dt=dt.astype({"Gender":'int',"Married":'int',"Dependents":'int',"Education":'int',"Self_Employed":'int',"ApplicantIncome":'int',"CoapplicantIncome":'int',"LoanAmount":'int',"Loan_Amount_Term":'int',"Credit_History":'int',"Property_Area":'int'})
```
